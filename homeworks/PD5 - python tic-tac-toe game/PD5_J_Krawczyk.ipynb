### Napisz gre "kółko i krzyżyk"

Wyświetlaj aktualny stan planszy przy pomocy znaków 'o', 'x', '-' w trzech wierszach
Np.:

```
--x
-x-
oo-
```

Naprzemiennie pytaj gdzie postawić 'x', a gdzie 'o' przyjmując numer pozycji od 1 do 9
Tzn.:

```
123
456
789
```

Zakończ grę gdy jedna ze stron wygra, lub wszystkie miejsca będą zapełnione

Na razie wyświetlaj wszystko po kolei, jedno pod drugim

Przykładowe wyjście

```
Kołko i krzyżyk!

---
---
---

Gdzie postawić 'x': 1

x--
---
---

Gdzie postawić 'o': 5

x--
-o-
---
```

I tak dalej

````python
plan = {1:"-", 2:"-", 3:"-", 4:"-", 5:"-", 6:"-", 7:"-", 8:"-", 9:"-"}
````

````python
def board():
    print(plan[1]+plan[2]+plan[3])
    print(plan[4]+plan[5]+plan[6])
    print(plan[7]+plan[8]+plan[9])
````

````python
def plan_update(number, setting):
    plan[number] = setting
    board()
````

````python
def plan_check():
    letters = ["o", "x"]
    winner = False
    for i in letters:
        if plan[1] == i and plan[4] == plan[1] and plan[7] == plan[1]:
            winner = True
        elif plan[1] == i and plan[2] == plan[1] and plan[3] == plan[1]:
            winner = True
        elif plan[1] == i and plan[5] == plan[1] and plan[9] == plan[1]:
            winner = True
        elif plan[2] == i and plan[5] == plan[2] and plan[8] == plan[2]:
            winner = True
        elif plan[5] == i and plan[4] == plan[5] and plan[6] == plan[5]:
            winner = True
        elif plan[8] == i and plan[7] == plan[8] and plan[9] == plan[8]:
            winner = True
        elif plan[3] == i and plan[6] == plan[3] and plan[9] == plan[3]:
            winner = True
        elif plan[3] == i and plan[5] == plan[3] and plan[7] == plan[3]: 
            winner = True
        else:
            pass
        if winner == True:
            print (f"wygrywa \"{i}\" ! gratulacje.")
            return True

#zwraca None albo print(wygrywa x/o)+True
````

````python
def repetitor():
    repeat = True
    while repeat == True:
        text = input("czy chcesz rozpocząć nową grę? (wpisz 't' jeśli tak, 'n' jeśli nie): ")
        if text not in "NnTt":
            print ("wprowadziłeś niedozwolony znak, spróbuj ponownie.")
        elif text in "Tt":
            print ("no to gramy!")
            global plan
            plan = {1:"-", 2:"-", 3:"-", 4:"-", 5:"-", 6:"-", 7:"-", 8:"-", 9:"-"}
            board()
            break
        elif text in "Nn":
            print ("do zobaczenia!")
            repeat = False
        else:
            pass
    if repeat == False:
        return False

# dla "n" printuje do zobaczenia i zwraca False (True dla False==repetitor())
# dla "g" powtarza pytanie
# dla "t" printuje pustą planszę i zwraca None (False dla False==repetitor())
````

````python
def play_check(play):
    checker = False
    if len(play) >= 2:
        print ("wprowadź maksymalnie jeden znak")
        checker = True
    if checker == True:
        return False
    if str(play) not in "123456789":
        print ("wpisuj tylko cyfry 1-9")
        checker = True
    if checker == True:
        return False
    if str(play) in "123456789" and plan[int(play)] != "-":
        print ("pomyliłeś się czy próbujesz coś kombinować? \npisz tylko w pustych miejscach!")
        checker = True
    if checker == True:
        return False

# dla cyfr 0-9 zwraca None
# dla wielku znaków printuje komunikat i zwraca False
# dla "0" i "g" printuje 'wpisuj tylko cyfry 1-9' i zwraca False
# dla cyfry, która już występowała na planszy printuje nie kombinuj i zwraca False
````

````python
def game():
    while plan_check() is not True:
        print ("1. grę rozpoczyna \"o\", podaj nr pola: ")
        play1 = input("gra \"o\", podaj nr pola: ")
        play_check(play1)
        if play_check(play1) == False:
            break
        plan_update(int(play1),"o")
        
        print ("2. dobry początek, teraz ruch \"x\":")
        play2 = input("gra \"x\", podaj nr pola: ")
        play_check(play2)
        if play_check(play2) == False:
            break
        plan_update(int(play2),"x")
        
        print ("3. ok, ruch \"o\":")
        play3 = input("gra \"o\", podaj nr pola: ")
        play_check(play3)
        if play_check(play3) == False:
            break
        plan_update(int(play3),"o")
        
        print ("4. pierwszy blok, co? ruch \"x\":")
        play4 = input("gra \"x\", podaj nr pola: ")
        play_check(play4)
        if play_check(play4) == False:
            break
        plan_update(int(play4),"x")
        
        print ("5. i wzajemnie. ruch \"o\":")
        play5 = input("gra \"o\", podaj nr pola: ")
        play_check(play5)
        if play_check(play5) == False:
            break
        plan_update(int(play5),"o")
        plan_check()
        if True == plan_check():
            repetitor()
        
        else:
            print ("6. tylko na tyle Cię stać? ruch \"x\":")
            play6 = input("gra \"x\", podaj nr pola: ")
            play_check(play6)
            if play_check(play6) == False:
                break
            plan_update(int(play6),"x")
            plan_check()
            if plan_check() == True:
                repetitor()
            
            else:
                print ("7. już prawie go masz. ruch \"o\":")
                play7 = input("gra \"o\", podaj nr pola: ")
                play_check(play7)
                if play_check(play7) == False:
                    break
                plan_update(int(play7),"o")
                plan_check()
                if plan_check() == True:
                    repetitor()
                
                else:
                    print ("8. zmyłka. ruch \"x\":")
                    play8 = input("gra \"x\", podaj nr pola: ")
                    play_check(play8)
                    if play_check(play8) == False:
                        break
                    plan_update(int(play8),"x")
                    plan_check()
                    if plan_check() == True:
                        repetitor()
                    
                    else:            
                        print ("9. na do widzenia: ruch \"o\":")
                        play9 = input("gra \"o\", podaj nr pola: ")
                        play_check(play9)
                        if play_check(play9) == False:
                            break
                        plan_update(int(play9),"o")
                        plan_check()
                        if plan_check() != True:
                            print ("no to mamy remis")
                            break
                            repetitor()
                        elif plan_check() == True:
                            repetitor()   
````

````python
print("KÓŁKO I KRZYŻYK \nInstrukcja do gry:\nwstaw swój znak (odpowiednio \"x\" lub \"o\") na planszy:")
print("123\n456\n789")
print("wpisując odpowiednią cyfrę w polu formularza.")
print("Ustaw trzy jednakowe znaki w linii zanim zrobi to Twój przeciwnik.\nPowodzenia!")

while repetitor() != False:
    game()
````
